## Распределение Лапласа

```python
mu = 2.0
sigma = 0.5

lapl = sts.laplace(loc=mu, scale=sigma)
vib = lapl.rvs(size=1000)
```

$\mu = 2$  
$b = 0.5$    
**vib** - выборка объёма 1000

```python
x = np.linspace(0,4,1000)
plt.hist(vib, normed=True, bins=100, label='Гистограмма выборки')
plt.plot(x, lapl.pdf(x), label='ТПР', alpha=sigma)
plt.legend()
plt.ylabel('$PDF&f(x)$')
plt.xlabel('$x$')
```

Гистограмма выборки  
ТПР - теоретическая плотность распределения

```python
z = {5:[],10:[],50:[],500:[]}
for i in 5,10,50,500:
    for b in range(1000):
        z[i].append(np.mean(lapl.rvs(size=i)))
```

$\textit{"Для этого при трёх и более значениях n сгенерируйте 1000 выборок объёма n"}$  

**z** - словарь хранящий в себе результаты расчетов выборочных средних для каждого n. 

Формула выборочного среднего:
$\bar{X}=\frac{1}{n}\sum_{i=1}^{n}x_{i}$  
Формула дисперсии для распределения Лапласа: $\sigma^2 = 2b^2$  
Среднеквадратическое отклонение: $\sigma = \sqrt{2b^2}$  
Математическое ожидание для распределения Лапласа: $\mu$

[Wiki - Laplace distribution](https://en.wikipedia.org/wiki/Laplace_distribution)

```python
plt.hist(z[5], label='n = 5', bins=30, normed=True)
norm = sts.norm(loc=mu, scale=np.sqrt(2*sigma**2/5))
plt.plot(x, norm.pdf(x), alpha=sigma)
plt.legend()
plt.ylabel('$PDF&f(x)$')
plt.xlabel('$x$')
```

**n** = 5  
Среднеквадратическое отклонение: $\sigma = \sqrt{\frac{2b^2}{n}} = \sqrt{\frac{2*0.5^2}{5}}$  
Математическое ожидание: $\mu = 2$ 

```python
plt.hist(z[10], label='Гистограмма n = 10', bins=30,  normed=True)
norm = sts.norm(loc=mu, scale=np.sqrt(2*(sigma**2)/10))
plt.plot(x, norm.pdf(x), alpha=sigma)
plt.legend()
plt.ylabel('$PDF&f(x)$')
plt.xlabel('$x$')
```

**n** = 10  
Среднеквадратическое отклонение: $\sigma = \sqrt{\frac{2b^2}{n}} = \sqrt{\frac{2*0.5^2}{10}}$  
Математическое ожидание: $\mu = 2$ 

```python
plt.hist(z[50], label='Гистограмма n = 50', bins=30,  normed=True)
norm = sts.norm(loc=mu, scale=np.sqrt(2*(sigma**2)/50))
plt.plot(x, norm.pdf(x), label='', alpha=sigma)
plt.legend()
plt.ylabel('$PDF&f(x)$')
plt.xlabel('$x$')
```

**n** = 50  
Среднеквадратическое отклонение: $\sigma = \sqrt{\frac{2b^2}{n}} = \sqrt{\frac{2*0.5^2}{50}}$  
Математическое ожидание: $\mu = 2$ 

```python
plt.hist(z[500], label='Гистограмма n = 500', bins=30,  normed=True)
norm = sts.norm(loc=mu, scale=np.sqrt(2*(sigma**2)/500))
plt.plot(x, norm.pdf(x), label='', alpha=sigma)
plt.legend()
plt.ylabel('$PDF&f(x)$')
plt.xlabel('$x$')
```

**n** = 500

Среднеквадратическое отклонение: $\sigma = \sqrt{\frac{2b^2}{n}} = \sqrt{\frac{2*0.5^2}{500}}$  
Математическое ожидание: $\mu = 2$ 

**Вывод: **Чем больше **n**, тем точнее нормальная аппроксимация, тем лучшее распределение описывается нормальным.
